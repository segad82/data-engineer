```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
markers = {'Dinner': 'D', 'Lunch': 's'}
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers)
plt.legend(loc='center', bbox_to_anchor=(1.11, 0.5))
plt.show()
```

```python
sns.lineplot(data=tips, x='total_bill', y='tip', hue='sex', style='time', size='size')
plt.legend(loc='center', bbox_to_anchor=(1.11, 0.5))
plt.show()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', kind='line')
plt.show()
```

```python
markers = {'Dinner': 'D', 'Lunch': 's'}
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', kind='scatter', col='sex', markers=markers)
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex')
plt.show()
```
