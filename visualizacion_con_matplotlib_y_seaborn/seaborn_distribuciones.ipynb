```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.histplot(data=tips, x='tip', bins=10)
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=True)
plt.show()
```

```python
# stat = [density, probability, percent, count, frequency]
sns.histplot(data=tips, x='tip', hue='sex', bins=15, cumulative=False, stat='frequency')
plt.show()
```

```python
# multiple = [dodge, stack, layer, fill]
sns.histplot(data=tips, x='tip', hue='sex', bins=15, cumulative=False, stat='count', multiple='dodge')
plt.show()
```

```python
# Diagrama de dencidad
sns.kdeplot(data=tips, x='tip', hue='sex')
plt.show()
```

```python
# La propiedad [shade] ahora es [fill]
sns.kdeplot(data=tips, x='tip', hue='sex', fill=True, bw_adjust=1)
plt.show()
```

```python
sns.displot(data=tips, x='tip', hue='smoker', kind='hist', multiple='dodge', stat='count')
plt.show()
```
