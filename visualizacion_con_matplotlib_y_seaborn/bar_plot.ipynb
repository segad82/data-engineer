```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['INDIA', 'JAPAN', 'MEXICO', 'COLOMBIA', 'GERMANY']
population = [1000,800,900,1000,300]
```

```python
plt.bar(country, population, color=['red', 'blue'], width=0.5)
plt.xticks(np.arange(5), ('India', 'Japón', 'Mexico', 'Colombia', 'Alemania'), rotation=45)
plt.xlabel('país')
plt.ylabel('n° habitantes')
plt.title('Cantidad de personas por país')
plt.show()
```

```python
plt.barh(country, population)
plt.show()
```
