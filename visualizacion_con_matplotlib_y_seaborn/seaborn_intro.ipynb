```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Seaborn
Está construido sobre Matplotlib e integrado para estructuras de Pandas.

#### Set
Pemite configurar los parámetros de visualización.

```python
sns.set(style='dark', palette='dark', font_scale=1, font="sans-serif")
sns.barplot(x=['A','B','C'], y=[3,5,7])
plt.show()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.displot( data=tips, x='total_bill')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', legend=False)
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', palette='pastel')
plt.show()
```
