```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
```

### Histograma

Ãštil para analizar la frecuencia de los datos.

```python
fig, axis = plt.subplots(nrows=1, ncols=3, figsize=(15,5))
axis[0].hist(data)
axis[1].hist(data, bins=50)
axis[2].hist(data, bins=50, histtype='step')
plt.show()
```

### Boxplot

Ãštil para analizar la distribuciÃ³n de los datos.

```python
fig, axis = plt.subplots(nrows=2, ncols=3, figsize=(12,5))
axis[0,0].boxplot(data)
axis[0,1].boxplot(data, vert=False)
axis[0,2].boxplot(data, patch_artist=True)
axis[1,0].boxplot(data, notch=True)
data = np.append(data, 100)
axis[1,1].boxplot(data, vert=False)
axis[1,2].boxplot(data, vert=False, showfliers=False)
plt.show()
```

### DispersiÃ³n (Scatter)

Ãštil para analizar relaciÃ³n entre variables.

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N)) ** 2
colors = np.random.rand(N)
```

```python
fig, axis = plt.subplots(nrows=2, ncols=3, figsize=(10,8))
axis[0,0].scatter(x,y)
axis[0,1].scatter(x,y, s=area)
axis[1,0].scatter(x,y, c=colors)
axis[1,1].scatter(x,y, s=area, c=colors, alpha=0.5)
axis[0,2].scatter(x,y, s=area, c=colors, alpha=0.5, marker='p')
plt.show()
```
