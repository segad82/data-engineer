## Matlab

```python
fig, axes = plt.subplots(figsize=(10,10))

axes.plot(x, x+1, 'ro--')
axes.plot(x, x+2, 'bD-')
axes.plot(x, x+3, 'y>-.')
axes.plot(x, x+4, 'g^:')
```

## Pyplot

```python
fig, axes = plt.subplots(figsize=(10,10))

axes.plot(x, x+1, color='green', linewidth=4, linestyle='dashed', marker='o')
axes.plot(x, x+2, color='yellow', linewidth=6, linestyle=':', marker='X')
axes.plot(x, x+3, color='#66FF66', alpha=0.6, linewidth=8, linestyle='--', marker='P', markersize=20, markerfacecolor='yellow')
axes.plot(x, x+4, color='#CCFFCC', linewidth=10, linestyle='-.', marker='8', markersize=20, markerfacecolor='red')
```
