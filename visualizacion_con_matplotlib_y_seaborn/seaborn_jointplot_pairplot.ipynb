```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.jointplot(data=tips, x='tip', y='total_bill', kind='kde')
plt.show()
```

```python
sns.jointplot(data=tips,
              x='tip',
              y='total_bill',
              hue='sex',
              kind='hist',
              marginal_ticks=True,
              marginal_kws=dict(bins=25, fill=False, multiple='dodge'))
plt.show()
```

```python
sns.pairplot(data=tips)
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', corner=True, palette='coolwarm')
plt.show()
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```
