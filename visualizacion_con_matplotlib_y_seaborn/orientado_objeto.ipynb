```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
layout = plt.figure()

axes1 = layout.add_axes([0.1,0.1,0.6,0.8])
axes2 = layout.add_axes([0.2,0.6,0.3,0.2])

axes1.plot(x,y)
axes1.set_title('Ejemplo')
axes2.plot(y,x)
axes2.set_facecolor('lightgrey')
layout.show()
```

Suplots

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(10,5))

axes[0,0].plot(x,np.sin(x), 'bo:', label='$sin(X)$')
axes[0,0].set_title('Relacin X - $sin(X)$')
axes[0,0].set_xlabel('X')
axes[0,0].set_ylabel('$sin(X)$')
axes[0,0].legend()

axes[0,1].plot(x,np.cos(x), 'ro:', label='$cos(X)$')
axes[0,1].set_title('Relacin X - $cos(X)$')
axes[0,1].set_xlabel('X')
axes[0,1].set_ylabel('$cos(X)$')
axes[0,1].legend()

axes[0,2].plot(x,np.tan(x), 'go:', label='$tan(X)$')
axes[0,2].set_title('Relacin X - $tan(X)$')
axes[0,2].set_xlabel('X')
axes[0,2].set_ylabel('$tan(X)$')
axes[0,2].legend()

axes[1,0].plot(y,np.sin(y), 'bo:', label='$sin(Y)$')
axes[1,0].set_title('Relacin Y - $sin(Y)$')
axes[1,0].set_xlabel('Y')
axes[1,0].set_ylabel('$sin(Y)$')
axes[1,0].legend()

axes[1,1].plot(y,np.cos(y), 'ro:', label='$cos(Y)$')
axes[1,1].set_title('Relacin Y - $cos(Y)$')
axes[1,1].set_xlabel('Y')
axes[1,1].set_ylabel('$cos(Y)$')
axes[1,1].legend(loc='lower right', bbox_to_anchor=(1.2,0.1))

axes[1,2].plot(y,np.tan(y), 'go:', label='$tan(Y)$')
axes[1,2].set_title('Relacin Y - $tan(Y)$')
axes[1,2].set_xlabel('Y')
axes[1,2].set_ylabel('$tan(Y)$')
axes[1,2].legend()

fig.tight_layout()
fig.set_facecolor('lightgrey')
```
