```python
import pandas as pd
input = 'sample_data.csv'
df = pd.read_csv(input)
df.head()
```

#### Pivot
La función se llama "pivot_table" y funciona igual que las tablas dinámicas de Excel.

```python
pv_df = df.pivot_table(index='Year', columns='Direction', values='Measure', aggfunc='count')
pv_df
```

#### Melt
La función melt sirve para tranformar columnas como filas y mantener segunda columna con valor asociado.

```python
df[['Year', 'Weekday']].head(3).melt()
```

```python
df.head(3).melt(id_vars='Year', value_vars='Weekday')
```
